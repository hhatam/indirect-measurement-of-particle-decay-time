```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
plt.rcParams["text.usetex"] = True
plt.rcParams["xtick.labelsize"] = 12
plt.rcParams["ytick.labelsize"] = 12
plt.rcParams["axes.labelsize"] = 16
plt.rcParams["font.family"] = "serif"
plt.rcParams["font.serif"] = "Computer Modern"
```

# $J/\psi$ Meson
## Loading the Data

```python
dataset = pd.read_csv('Jpsimumu_Run2011A.csv')
masses = dataset['M']
```

## Plotting

```python
fig, ax = plt.subplots()
ax.set_xlabel("Mass (GeV)")
ax.set_ylabel("Events")
ax.hist(masses, bins=500, color="mediumblue")
plt.savefig("jpsi_full_mass_hist.pdf", bbox_inches="tight")
```

```python
peak_masses = masses[(masses > 3.0) & (masses < 3.2)]

fig, ax = plt.subplots()
ax.set_xlabel("Mass (GeV)")
ax.set_ylabel("Events")

ax.hist(peak_masses, bins=500, density=True, color="mediumblue")

mean_mass = np.mean(peak_masses)
var_mass = np.var(peak_masses)

x = np.linspace(3.0, 3.2, 10000)
y = np.exp(-0.5*(x-mean_mass)**2/var_mass) / np.sqrt(2*np.pi*var_mass)
ax.plot(x, y, color="red", lw=3)

plt.savefig("jpsi_peak_mass_hist.pdf", bbox_inches="tight")
```

# Calculations

```python
std_mass = np.std(peak_masses)
print("Mean Mass:", mean_mass, "GeV\nStandard Deviation of the Mass:", std_mass, "GeV")
```

```python
hbar = 1.054571817e-34
e = 1.6021766208e-19
lifetime = hbar / (2 * std_mass * e * 1e9)

print("Lifetime:", lifetime, "s")
```
